```python
class N:
    def __init__(self,l,r,freq):
        self.l = l
        self.r = r
        self.freq = freq
    
    def get_freq(self):
        return self.freq
        
class L:
    def __init__(self,le,lefreq):
        self.le =le
        self.lefreq =lefreq
        #self.code_word = None
        
    def get_freq(self):
        return self.lefreq
        


def histogram(a):
#     a = a.lower()
    freqs ={} # counts
    
    for i in a:
        freqs[i]= freqs.get(i,0)+1

    return freqs


def func(x):
        return x.get_freq()


def sh(freqs):
    sort_histo=[] # list of objects
    for i,j in freqs.items():
        sort_histo.append(L(i,j))
        
    sort_histo = sorted(sort_histo,key =func) #[L1,l2,l3]
    
    for i in sort_histo:
        print(i.le+":"+str(i.lefreq))
    return sort_histo
           
# make a tree

def tree(sort_histo):
    while(len(sort_histo)>1):
        
        m1 = min(sort_histo,key=func)  # m1 is a object which points to a minimum obj in sort_histo
        sort_histo.remove(m1)          # minimum leaf 1 so 
        
        m2 = min(sort_histo,key=func)
        sort_histo.remove(m2)          # minimum leaf 2
        
        nm = m1.get_freq()+m2.get_freq()  # get that leaf1.get_freq , leaf1 is an obj and get_freq will give the leaf's frequency
        
        t1 = N(m1,m2,nm)
         
        sort_histo.append(t1)
    return sort_histo            # returning a tree  [l1 ,l2 ,t1]
        
```

```python
def inorder(r, code,enco):
    if r:
        if type(r).__name__ == 'N':
            inorder(r.l, code + '0',enco)
            #print(r.get_freq())
        if type(r).__name__ == 'L':
            #r.code_word = code
            #print(r.le + ':' + r.code_word)
            enco.append(tuple([r.le,code]))
            print(r.le+":"+code)
        if type(r).__name__ == 'N':
            inorder(r.r, code + '1',enco)
```

```python
def encoding(a,enco):
    encol=[]
    for i in a:
        for j in enco:
            if i == j[0]:
                encol.append(j[1])
                break
    return "".join(encol)
```

```python
def decoding(enco,encoded):
    buff=""
    final =""
    for i in encoded:
        buff= buff+i
        for j in enco:
            if buff== j[1]:
                final= final+j[0]
                encoded = encoded.replace(buff,"")
                buff=""
    print(final)
                      
```

```python
 a = input("Enter a text: ")
# a = 'maggieag'
print("---------------------------------------------")

print("Frequency of each character:")
_tree = sh(histogram(a))  # _tree is a list of objects 
print("---------------------------------------------")


#print(_tree[0].le) will print first letter
print("Codes of each character:")
fin_tree = tree(_tree)[0]
enco=[]
inorder(fin_tree, '',enco)
print("---------------------------------------------")


print(encoding(a,enco))  #1101000100111
encoded = encoding(a,enco)  # encoded is string
# print(encoded)
print("---------------------------------------------")

print("The text is :")
decoding(enco,encoded)
```
